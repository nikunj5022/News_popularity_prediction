```python
#importing required libraries
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from news_crawling import required_details
```

```python
#loading required columns only
cols = [' n_tokens_title',' n_tokens_content',' n_unique_tokens',' n_non_stop_unique_tokens',
        ' num_imgs',' average_token_length',' weekday_is_monday',' weekday_is_tuesday',' weekday_is_wednesday',
        ' weekday_is_thursday',' weekday_is_friday',' weekday_is_saturday',' weekday_is_sunday',' is_weekend',
        ' global_subjectivity',' global_sentiment_polarity',' title_subjectivity',' title_sentiment_polarity',' shares']
data = pd.read_csv('C:/Users/My HP Pavilion/Desktop/OnlineNewsPopularity/OnlineNewsPopularity.csv',usecols=cols)
```

```python
data.head()
```

```python
data.describe()
```

```python
#removing outliers
q = data[' n_unique_tokens'].quantile(0.999)
data_1 = data[data[' n_unique_tokens']<q]

q = data_1[' num_imgs'].quantile(0.99)
data_2 = data_1[data_1[' num_imgs']<q]

q = data_2[' shares'].quantile(0.99)
data_3 = data_2[data_2[' shares']<q]

q = data_3[' average_token_length'].quantile(0.02)
data_4 = data_3[data_3[' average_token_length']>q]

q = data_4[' shares'].quantile(0.01)
data_5 = data_4[data_4[' shares']>q]

data_final = data_5
```

```python
data_final[[' n_tokens_title',' n_tokens_content',' n_unique_tokens',
        ' n_non_stop_unique_tokens',' num_imgs',' average_token_length']] = StandardScaler().fit_transform(data_final[[' n_tokens_title',' n_tokens_content',' n_unique_tokens',
        ' n_non_stop_unique_tokens',' num_imgs',' average_token_length']])
data_final.head()
```

```python
data = data_final.drop([' shares'],axis=1)
output = data_final[' shares']
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, output, test_size=0.15, random_state=101)
```

```python
#regression
regression_model = LinearRegression()
regression_model.fit(x_train, y_train)
```

```python
predict_regression_output = regression_model.predict(x_test)
```

```python
print('Predicted output',predict_regression_output[:5])
print('Actual output',list(y_test[:5]))
```

```python
#random forest
random_forest = RandomForestRegressor(random_state=42)
random_forest.fit(x_train, y_train)
```

```python
predict = random_forest.predict(x_test)
print('Predicted output',predict[34:40])
print('Actual output',list(y_test[34:40]))
```
